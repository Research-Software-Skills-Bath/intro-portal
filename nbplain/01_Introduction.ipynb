## What is the Research Computing Account Management Portal?

The Research Computing Account Management (RCAM) Portal allows users to manage their research computing resources and finance.

The portal allows users to:

- create and manage accounts, membership, and resources
- view job transactions, storage consumption, budgets, and expenditure.

You will need to use the Research Computing Account Management Portal if you need to:

- Set up and configure an account to use research computing resources such as Nimbus and Anatra.
- Monitor the budget commitment and expenditure associated with a Research Computing account.
- Monitor computing and storage transactions.

# Accounts and Resource Allocations

The RCAM Portal is structured as a hierarchical relationship between accounts and resource allocations. This 2-level account and resource allocation hierarchy allows the budget associated with an Agresso project to be apportioned across multiple resource allocations with different members, active periods, resources, and resource configurations etc.

![Accounts Resource Allocations](../images/system-relationships.png)


- **Accounts**:
At the top level, accounts have a one-to-one relationship with an Agresso project, and at this level, users can use the portal to view the account budget and expenditure, allocate budget across resource allocations and manage account membership.

- **Resource Allocations**:
Each account can subsequently have one or many resource allocations and at this level, users can assign, and configure compute and storage resources, view job transactions and storage consumption, and manage the resource allocation membership.
