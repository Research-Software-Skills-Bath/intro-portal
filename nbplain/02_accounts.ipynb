## Where can I see the accounts I have access to?

You can navigate to the accounts screen to see what accounts you have on the system from the portal home page:

![Research Finance Portal](../images/ResearchFinancePortalAccount.png)

or from the sidebar `Accounts` menu:

![Research Finance Portal](../images/AccountSidebar.png)


Once you’ve selected Accounts you will see a list of accounts you have access to and against each account listed a set of options for Managing/ Viewing account, allocating budgets, viewing finance, and viewing storage will be available.

## How do I add an account?

Navigate to the accounts screen and click the `set up new account` button:

![Portal Account](../images/SetUpAccount1.png)

From the drop down menu in `Link Agresso Project Code` select the Agresso code you wish to set up a new account for.

![Portal Account](../images/SetUpAccount2.png)

![Portal Account](../images/SetUpAccount3.png)


If you are expecting to see an Agresso project code in the list but it is not shown, it is either because:

- The Agresso project code has already been linked to an account
- You have not been set up as the budget holder in Agresso
- The Agresso project code has not been allocated a budget for High Performance Computing

Information on requesting a budget for High Performance Computing can be found [here](https://rcam.bath.ac.uk/HPC%20Budget%20Request%20Form%20Guide.pdf).

Once you have selected the Agresso account you wish to set up click `Create and Next`

![Portal Account](../images/SetUpAccount4.png)

The next screen simply lists the Agresso Account details:

![Account Set up](../images/SetUpAccount5.png)

On the next Screen you can add a description and tags to the account:

![Account Set up](../images/SetUpAccount6.png)

You can then assign the **Account Administrator** roll to other users. This role grants the member access to manage the account with the ability to:

- Allocate budgets across resource allocations
- View budget and expenditure
- Manage **all** resource allocations within the account

The budget holder is automatically assigned as an Account Administrator. To assign this role to another person, type the name or the username in the search box and click add member:

![Account Set up](../images/SetUpAccount7.png)

You can then assign the **Account Viewer** role to other users. This role gives the user view access to the account including the ability to:

- View access to the budget allocations and finance
- View access to all the resource allocations 
- View access to all resource allocation transactions within the account. 

To assign this role to another person, type the name or the username in the search box and click add member:

![Account Set up](../images/SetUpAccount8.png)

## Account Storage (Account-level quotas)

Account Storage allows you to view the storage quota associated with an account and monitor the latest usage.

**Questions**
- Where do I view Account Storage?
- What do the quota and usage fields mean?
- Who can change storage quotas?


## Where do I view Account Storage?

1. Go to **Accounts** from the left sidebar.
2. Locate the account you want to view.
3. In the **Actions** column, select the **Storage Allocations** icon.

![Accounts – Storage Allocations](../images/Account_Storage_01.png)

This opens the **Account Storage** page for the selected account.

![Account Storage Page](../images/Account_Storage_02.png)


## Adding storage to an account

On the **Account Storage** page:

1. Select a storage option from **“Please select storage option…”**.
2. Click **Add**.

**Permissions**
- Only **Research Account Administrators** can set or change storage quotas.
- If you are **Research Account Viewer**, you may still be able to *view* quotas and usage, but you must contact the Research Computing team to request changes.


## Understanding the storage table

Each row represents a storage allocation attached to the account (for example, **Nimbus Campaign**).

- **Total Account Allocation (GB):** The quota assigned at the account level.
- **Total Quota Across Resource Allocations (GB):** The total quota distributed across allocations under the account.
- **Total Latest Usage (GB):** Latest reported usage (updated periodically during the day).


## Editing an existing storage quota (Admins only)

If you have administrator permissions, use the **Edit (pencil)** action on the right-hand side of the storage row to adjust the quota.

If you do not have permissions, contact the Research Computing team and include:

- The **Account ID**
- The **storage name** (e.g., “Nimbus Campaign”)
- The **requested quota (GB)** and a brief reason
