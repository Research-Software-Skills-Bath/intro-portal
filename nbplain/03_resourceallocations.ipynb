# Resource Allocations

In order to to use the clusters, and actually _submit a job_ you will need to associate a resource allocation with an account. Each Agresso account can have multiple resource allocations. 

In the Resource Allocation admins can:
- Assign and configure compute and storage resources
- View job transactions and storage consumption
- Manage the resource allocation membership

## How do I add a Resource Allocation?

Navigate to the resource allocation screen and click the `create new resource allocation` button:

![Portal Account](../images/SetUpRA1.png)

From the select account page pick which of your Agresso accounts you wish to set up a Resource Allocation for:

![Portal Account](../images/SetUpRA2.png)

Now you can fill in details of the RA, inlcuding the title, description, the status and set start and end dates:

![Portal Account](../images/SetUpRA3.png)

Next you can grant the *Resource Allocation Administrator* role to users: This gives full access to manage the resource allocation allowing users the ability to:

- assign other members to the resource allocation
- add/remove, and configuring resources
- submitting jobs against this resource allocation in Slurm 
- viewing the job and storage transactions against the resource allocation. 

To assign this role to another person, type the name or the username in the search box and click add member:

![Portal Account](../images/SetUpRA4.png)

Next you can grant the Resource Allocation User role to users. This role will allow a user to:

- Submit jobs against a resource allocation on the cluster 
- View the resource allocation but make no changes to it


![Portal Account](../images/SetUpRA5.png)

Finally we can add a cluster to the resource allocation: 

![Portal Account](../images/SetUpRA6.png)
